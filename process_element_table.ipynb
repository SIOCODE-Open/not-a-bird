```python
# You need to do this first before running this notebook:
# python -m pip install PyYAML
import csv
import yaml

filename = 'unfires_element_table.csv'

itemsyaml = []
recipesyaml = []

with __builtins__.open(filename) as f:
    csvreader = csv.reader(f)
    csvheader = next(csvreader)
    elementslist = csvheader[1:]
    csvrows = list(csvreader)
    for y in range(len(csvrows)):
        combo = csvrows[y][1:]
        recipe = []
        for x in range(len(combo)):
            comboint = int(combo[x])
            if comboint > 0:
                for i in range(comboint):
                    recipe.append(elementslist[x])
        if len(recipe) == 0:
            print(f'{csvrows[y][0]}: ROOT ELEMENT (tier = 1)')
            itemsyaml.append({
                'name': csvrows[y][0], 
                'description': 'We need a description here',
                'tier': 1
            })
        elif len(recipe) == 2:
            print(f'{csvrows[y][0]}: {" + ".join(recipe)}')
            itemsyaml.append({
                'name': csvrows[y][0], 
                'description': 'We need a description here',
                'tier': 2,
            })
            recipesyaml.append({
                'name': 'Make ' + csvrows[y][0],
                'a': f'indexed-items#{recipe[0]}',
                'b': f'indexed-items#{recipe[1]}',
                'c': f'indexed-items#{csvrows[y][0]}',
            })
        else:
            print(f'{csvrows[y][0]}: INVALID RECIPE ({" + ".join(recipe)})')
            itemsyaml.append({
                'name': csvrows[y][0], 
                'description': 'We need a description here',
                'tier': 1
            })

with __builtins__.open('items.pdata.yaml', 'w') as f:
    yaml.dump({"objects": itemsyaml}, f)

with __builtins__.open('recipes.pdata.yaml', 'w') as f:
    yaml.dump({"objects": recipesyaml}, f)
```
