```python
def pascal_case(name):
    parts = name.split(' ')
    return ''.join(part.capitalize() for part in parts)

all_elements = []
for a, b, c in recipes:
    a_name = pascal_case(a)
    b_name = pascal_case(b)
    c_name = pascal_case(c)
    if a_name not in all_elements:
        all_elements.append(a_name)
    if b_name not in all_elements:
        all_elements.append(b_name)
    if c_name not in all_elements:
        all_elements.append(c_name)
for element in all_elements:
    print(f"  - name: {element}")
    print(f"    description: We need a description here")
    print(f"    tier: 1")
```
